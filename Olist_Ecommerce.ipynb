```python
#MONTHLY REVENUE USING SQL

spark.sql("""
SELECT
date_format(o.order_purchase_timestamp,'yyyy-MM') as month,
round(sum(p.payment_value),2) as revenue
FROM orders o
JOIN payments p
ON o.order_id=p.order_id
GROUP BY month
ORDER BY month
""").show()
```

```python
# Top 10 customers

spark.sql("""
SELECT
c.customer_unique_id,
round(sum(p.payment_value),2) revenue
FROM customers c
JOIN orders o
ON c.customer_id=o.customer_id
JOIN payments p
ON o.order_id=p.order_id
GROUP BY customer_unique_id
ORDER BY revenue DESC
LIMIT 10
""").show()
```

```python
# Top categories

spark.sql("""
SELECT
t.product_category_name_english,
round(sum(p.payment_value),2) revenue
FROM order_items oi
JOIN products pr
ON oi.product_id=pr.product_id
JOIN translation t
ON pr.product_category_name=t.product_category_name
JOIN payments p
ON oi.order_id=p.order_id
GROUP BY product_category_name_english
ORDER BY revenue DESC
LIMIT 10
""").show()
```

```python
# Window Function

from pyspark.sql.window import Window
from pyspark.sql.functions import rank,sum, dense_rank
```

```python
customer_revenue = (
    customers
    .join(orders,"customer_id")
    .join(payments,"order_id")
    .groupBy("customer_unique_id")
    .agg(sum("payment_value").alias("revenue"))
)


window_spec = Window.orderBy(
    customer_revenue.revenue.desc()
)
```

```python
ranked = customer_revenue.withColumn(
    "rank",
    rank().over(window_spec)
)

ranked.show()
```

```python
dense_ranked = customer_revenue.withColumn(
    "dense_rank",
    dense_rank().over(window_spec)
)

dense_ranked.show()
```

```python
monthly_revenue = spark.sql("""
SELECT
date_format(
order_purchase_timestamp,
'yyyy-MM'
) month,
sum(payment_value) revenue
FROM orders o
JOIN payments p
ON o.order_id=p.order_id
GROUP BY month
ORDER BY month
""")

monthly_revenue.show()
```

```python
window_month = Window.orderBy("month")
```

```python
monthly_revenue.withColumn(
    "running_revenue",
    sum("revenue").over(window_month)
).show()
```

```python
monthly_revenue.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("monthly_revenue")
```
